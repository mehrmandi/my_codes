```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('data\creditcard.csv')
data.head()
```

```python
np.unique(data['Class'], return_counts=True)
```

```python
from sklearn.neighbors import LocalOutlierFactor
```

```python
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.002, n_jobs=-1)
```

```python
X = data.drop('Class', axis=1)
```

```python
y_pred = lof.fit_predict(X)
```

```python
np.unique(y_pred, return_counts=True)
```

```python
y_pred[y_pred == -1] = 1
y_pred[y_pred == 1] = 0
```

```python
from sklearn import metrics
```

```python
y_true = data['Class']
```

```python
print(metrics.classification_report(y_true=y_true, y_pred=y_pred))
```

```python
from sklearn.ensemble import IsolationForest
```

```python
IF = IsolationForest(contamination=0.01,
                     random_state=42,
                     n_jobs=-1)
y_pred = IF.fit_predict(X)
```

```python
y_pred_if = np.where(y_pred == -1, 1, 0)
```

```python
print(metrics.classification_report(y_true=y_true, y_pred=y_pred_if))
```
